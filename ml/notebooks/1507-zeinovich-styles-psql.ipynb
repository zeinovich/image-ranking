```python
import pandas as pd
import numpy as np
import pickle
```

# Preparation

Column styles contains dict which is not suitable for PGSQL database

```python
styles = pd.read_pickle('../data/styles-76a25060.pkl')
styles.head()
```

```python
styles.shape
```

```python
images = pd.json_normalize(styles.images)
images.head()
```

```python
images = images.fillna('')
images.head()
```

```python
all(images.index == styles.index)
```

```python
merged = styles.merge(
            right=images,
            how='left',
            left_index=True,
            right_index=True,
            suffixes=('', ''),
            ).drop('images', axis=1)
```

```python
merged.head()
```

```python
merged = merged.reset_index(drop=False)
merged.to_pickle('../data/styles-76a25070.pkl')
```
